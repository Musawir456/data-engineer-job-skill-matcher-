# 2. Cleaning function:

```python
def clean_text(t):
    t = str(t).lower()
    t = re.sub(r"[^a-z0-9+.# ]", " ", t)
    t = re.sub(r"\s+", " ", t)
    return t.strip()

df["clean_desc"] = df["Job Description"].apply(clean_text)
df[["Job Title", "clean_desc"]].head()
```

```python
df.shape
```

```python
def extract_skills(text, skill_list):
    text = " " + text + " "
    found = []
    for s in skill_list:
        if " " + s + " " in text:
            found.append(s)
    return list(set(found))

df["job_skills"] = df["clean_desc"].apply(lambda x: extract_skills(x, core_skills))

df[["Job Title", "job_skills"]].head()
```
